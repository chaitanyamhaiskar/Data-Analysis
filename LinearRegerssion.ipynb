# After EDA
## we will now split the data set for training and testing

```python
from sklearn.model_selection import cross_val_score,KFold,cross_validate,train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.preprocessing import LabelEncoder
def labelEncoding(df,col):
  le = LabelEncoder()
  le.fit(df[col])
  df[col] = le.transform(df[col])
  print('Transformed Successfully',col)
```

```python
# labelEncoding(df,['Eduacation','Race','Hisp','MaritalStatus'])
labelEncoding(df,'MaritalStatus')
```

```python
df_num = df.select_dtypes(exclude = 'object')
df_cat = df.select_dtypes(include = 'object')
```

```python
x = df.drop(columns=['Earnings_1978'])
y = df['Earnings_1978']
```

```python
lr = LinearRegression()
```

```python
kfold = KFold(n_splits=10,shuffle=True,random_state=42)
```

```python
scores = cross_val_score(lr, x, y, cv=kfold, scoring='neg_mean_squared_error')
```

```python
mse_scores = -scores
mean_mse = np.mean(mse_scores)
std_mse = np.std(mse_scores)
print(mean_mse)
```

```python
print(f"Mean MSE: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
# print(r2_score(y_true,y_pred))
print(scores)
```

```python
# now will use train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr.fit(x_train,y_train)
ypred = lr.predict(x_test)
ypred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,ypred)
accuracy = r2_score(y_test,ypred)
accuracy
```
